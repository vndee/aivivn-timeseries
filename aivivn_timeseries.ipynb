```
import pandas as pd

data = pd.read_csv('data/train.csv')
print('Shape:', data.shape)
print(data.head(10))
```

```
day = data.groupby(['UPDATE_TIME']).mean()
day_bandwidth = day.loc[:, 'BANDWIDTH_TOTAL']
day_maxuser = day.loc[:, 'MAX_USER']

print('Shape:', day.shape)
print(day.head(10))
```

```
day_bandwidth_chart = day_bandwidth.plot.line(figsize=(20,4), title='Bandwidth per day')
```

```
day_maxuser_chart = day_maxuser.plot.line(figsize=(20,4), title='Max user per day')
```

```
hour = data.groupby(['HOUR_ID']).mean()
print('Shape:', hour.shape)
print(hour.head(5))
```

```
hour_bandwidth = hour.loc[:, 'BANDWIDTH_TOTAL'].plot.bar(figsize=(20, 4), title='Bandwidth per hour')
```

```
hour_bandwidth = hour.loc[:, 'MAX_USER'].plot.bar(figsize=(20, 4), title='Max user per hour')
```

```
zone = data.groupby(['ZONE_CODE']).mean()
print('Shape:', zone.shape)
print(zone.head(5))
```

```
zone_bandwidth = zone.loc[:, 'BANDWIDTH_TOTAL']
print('Shape:', zone_bandwidth.shape)
print(zone_bandwidth.head(5))
```

```
zone_bandwidth_chart = zone_bandwidth.plot.bar(title='Bandwidth per zone')
```

```
zone_maxuser = zone.loc[:, 'MAX_USER']
print('Shape:', zone_maxuser.shape)
print(zone_maxuser.head(5))
```

```
zone_maxuser_chart = zone_maxuser.plot.bar(title='Max user per zone')
```

```
server = data.groupby(['SERVER_NAME']).mean()
print('Shape:', server.shape)
print(server.head(5))
```

```
server_bandwidth = server.loc[:, 'BANDWIDTH_TOTAL']
print('Shape:', server_bandwidth.shape)
print(server_bandwidth.head(5))
```

```
import matplotlib.pyplot as plt

server_bandwidth_chart = server_bandwidth.plot.bar(title='Bandwidth per server', figsize=(20, 4))
ax1 = plt.axes()
x_axis = ax1.axes.get_xaxis()
x_axis.set_visible(False)
plt.show()
```

```
server_maxuser = server.loc[:, 'MAX_USER']
print('Shape:', server_maxuser.shape)
print(server_maxuser.head(5))
```

```
server_maxuser_chart = server_maxuser.plot.bar(title='Max user per server', figsize=(20, 4))
ax1 = plt.axes()
x_axis = ax1.axes.get_xaxis()
x_axis.set_visible(False)
plt.show()
```

```
server_chart = server.loc[:, 'BANDWIDTH_TOTAL'].plot.line(title='SERVER BANDWIDTH', figsize=(20, 4))
ax1 = plt.axes()
x_axis = ax1.axes.get_xaxis()
x_axis.set_visible(False)
plt.show()
```

```
server_chart = server.loc[:, 'MAX_USER'].plot.line(title='SERVER MAX USER', figsize=(20, 4))
ax1 = plt.axes()
x_axis = ax1.axes.get_xaxis()
x_axis.set_visible(False)
plt.show()
```
